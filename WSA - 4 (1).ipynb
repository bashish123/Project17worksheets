#### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
 #import require libraries
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
import time
import itertools
```

```python
#CONNECT TO WEB-DRIVER
driver=webdriver.Chrome(r"C:\Users\Ashish\Downloads\chromedriver.exe")
```

```python
# Step 1 :- 1. first get the webpage https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
#open the site in our automatic chrome browser
driver.get(url)
```

```python
Rank=[]
Name=[]
Artist=[]
U_D=[]
Views=[]

try:
    rank=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rank:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('- -')
except StaleElementReferenceException:
    Rank.append('- -')
#########################################################################

try:
    name=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('- -')
except StaleElementReferenceException:
    Name.append('- -')
#########################################################################

try:
    artist=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('- -')
except StaleElementReferenceException:
    Artist.append('- -')
#########################################################################

try:
    view=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('- -')
except StaleElementReferenceException:
    Views.append('- -')
#########################################################################

try:
    upload=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload:
        U_D.append(i.text)
except NoSuchElementException:
    U_D.append('- -')
except StaleElementReferenceException:
    U_D.append('- -')
```

```python
len(Rank),len(Name),len(Artist),len(Views),len(U_D)
```

```python
#CreateDataframe
df=pd.DataFrame()
df['Rank']=Rank
df['Name']=Name
df['Arist']=Artist
df['View']=Views
df['Upload Date']=U_D
df
```

```python
#Save as .csv
df.to_csv("YT_Videos")
```
